```python
import pandas as pd

# Load the dataset (adjust the file path if needed)
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
data = pd.read_csv(url)

# Show the first few rows
print(data.head())
```

```python
# Group by country/region and sum the cases
data_grouped = data.groupby('Country/Region').sum()

# Drop non-date columns (like 'Lat', 'Long', 'Country/Region')
data_grouped = data_grouped.drop(columns=['Lat', 'Long'])

# Show the results
print(data_grouped.head())
```

```python
import matplotlib.pyplot as plt

# Choose a country to visualize (e.g., United States)
country = 'US'
country_data = data_grouped.loc[country]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(country_data.index, country_data.values, label=f'Confirmed Cases - {country}')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title(f'COVID-19 Confirmed Cases Over Time - {country}')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
